```python
import pandas as pd
```

```python
def data_join (transactionFile, identityFile):
    transactions = pd.read_csv(transactionFile)
    identity = pd.read_csv(identityFile)
    joined = transactions.merge(identity, how='left', on='TransactionID')
    return joined
```

```python
trainTransactions = 'Data/train_transaction.csv'
trainIdentity = 'Data/train_identity.csv'
train_df = data_join(trainTransactions, trainIdentity)
train_df.head()
```
