```python
import pandas as pd
```

```python
def data_join (transactionFile, identityFile):
    transactions = pd.read_csv(transactionFile)
    identity = pd.read_csv(identityFile)
    joined = transactions.merge(identity, how='left', on='TransactionID')
    return joined
```

```python
testTransactions = 'Data/test_transaction.csv'
testIdentity = 'Data/test_identity.csv'
df = data_join(testTransactions, testIdentity)
df.head()
```
